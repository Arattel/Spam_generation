```python
import os
from tqdm import tqdm
from helpers.parsing import check_file_extension, extract_header_and_body, get_content_type, process_content_type
```

```python
files_data = []
dataset_directory = './data/'
file_extensions = ['.ivanova', '.lorien', '.delenn']
```

```python
content_types = []
```

```python
for root, dirs, files in tqdm(os.walk(dataset_directory)):
    for file in tqdm(files):
        path = os.path.join(root, file)
        if check_file_extension(file_extensions, file):
            header, body = None, None 
            with open(path, 'r', encoding="utf8", errors='ignore') as f:
                header, body = extract_header_and_body(f)
            content_types.append((process_content_type(get_content_type(header)), body))
```

```python
content_types_count  = {}
```

```python
for key, body in tqdm(content_types):
    if key in content_types_count:
        content_types_count[key] += 1
    else:
        content_types_count[key] = 1
```

```python
content_types_count
```

```python
keys = list(content_types_count.keys())
for key in keys:
    if  key is not None and '/' not in key:
        del content_types_count[key]
```

```python
content_types_count
```

```python
for content_type in content_types_count:
    print('-' * 38)
    print('TYPE: ', content_type)
    for message_type, message in content_types:
        if message_type == content_type:
            print(message)
            break
    print('-' * 38)
```

```python
content_types = list(filter(lambda x: x[0] is not None, content_types))
content_types = list(filter(lambda x: x[0].strip() == 'text/plain', content_types))
len(content_types)
```
